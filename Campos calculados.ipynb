# Ejemplos más utilizados:

```python
Para aplicar a cambios de colores:

// 1. Resaltar Ventas Altas
IF SUM([Sales]) > 10000 THEN "High Sales" ELSE "Low Sales" END

// 2. Resaltar Crecimiento Interanual Positivo
IF (SUM([Sales]) - LOOKUP(SUM([Sales]), -1)) > 0 THEN "Growth" ELSE "Decline" END

// 3. Resaltar Desempeño por Región
IF [Region] = "East" THEN "East Region" ELSE "Other Regions" END

// 4. Resaltar Categorías con Alta Rentabilidad
IF SUM([Profit]) > 5000 THEN "High Profit" ELSE "Low Profit" END

// 5. Resaltar Productos con Stock Bajo
IF SUM([Quantity]) < 50 THEN "Low Stock" ELSE "Sufficient Stock" END


Para datos usuales:

// Suma de Ventas
SUM([Sales])

// Promedio de Ventas
AVG([Sales])

// Valor Mínimo de Ventas
MIN([Sales])

// Valor Máximo de Ventas
MAX([Sales])

// Conteo de Registros
COUNT([Sales])

// Conteo de Valores Distintos de Clientes
COUNTD([Customer ID])

// Mediana de Ventas
MEDIAN([Sales])

// Desviación Estándar de Ventas (Muestra)
STDEV([Sales])

// Desviación Estándar de Ventas (Población)
STDEVP([Sales])

// Varianza de Ventas (Muestra)
VAR([Sales])

// Varianza de Ventas (Población)
VARP([Sales])

// Atributo (Un único valor si todos son iguales)
ATTR([Product Category])

// Agregación Personalizada
AGG([Custom Calculation])

// Valor Nulo como Cero
ZN([Sales])

// Diferencia desde el Valor Anterior
LOOKUP([Sales], -1)

// Porcentaje del Total de Ventas
SUM([Sales]) / TOTAL(SUM([Sales]))

// Ventas Acumuladas
RUNNING_SUM(SUM([Sales]))

// Promedio Móvil de Ventas
WINDOW_AVG(SUM([Sales]), -2, 0)

// Clasificación de Ventas
RANK(SUM([Sales]))

// Días desde la Primera Compra
DATEDIFF('day', MIN([Order Date]), TODAY())

// Concatenación de Nombres de Productos
CONCATENATE([Product Name], ", ")

// Extracción de Año de una Fecha
YEAR([Order Date])

// Si una Condición es Verdadera
IF [Sales] > 1000 THEN "High" ELSE "Low" END

// Convertir a String
STR([Sales])

// Extraer Subcadena
LEFT([Customer Name], 5)

// Convertir a Fecha
DATE([Order Date])

// Redondear al Entero más Cercano
ROUND([Sales])

// Convertir a Entero
INT([Sales])

// Capitalizar Texto
UPPER([Customer Name])

// Longitud de la Cadena
LEN([Customer Name])

// Encontrar Posición de Subcadena
FIND([Customer Name], "Smith")

// Reemplazar Subcadena
REPLACE([Customer Name], "Smith", "Jones")

// Fecha Actual
TODAY()

// Número de Ventas por Trimestre
DATETRUNC('quarter', [Order Date])


// Cálculo de Total hasta la Fecha (YTD)
TOTAL(SUM([Sales]))

// Porcentaje de Crecimiento Interanual
(SUM([Sales]) - LOOKUP(SUM([Sales]), -1)) / ABS(LOOKUP(SUM([Sales]), -1))

// Porcentaje de Diferencia desde el Valor Anterior
(SUM([Sales]) - LOOKUP(SUM([Sales]), -1)) / LOOKUP(SUM([Sales]), -1)

// Cálculo de Percentil
PERCENTILE([Sales], 0.9)

// Límite Superior del Percentil
WINDOW_MAX([Sales])

// Límite Inferior del Percentil
WINDOW_MIN([Sales])

// Diferencia de Ventas con respecto al Año Anterior
SUM([Sales]) - LOOKUP(SUM([Sales]), -1)

// Coeficiente de Correlación
CORR([Sales], [Profit])

// Coeficiente de Covarianza
COVAR([Sales], [Profit])

// Coeficiente de Determinación (R^2)
RSQ([Sales], [Profit])

// Concatenación Condicional
IF [Category] = "Furniture" THEN CONCATENATE([Product Name], " (F)") ELSE [Product Name] END

// Suma Condicional
IF [Region] = "East" THEN SUM([Sales]) END

// Promedio Condicional
IF [Region] = "West" THEN AVG([Sales]) END

// Redondeo hacia Abajo
FLOOR([Sales])

// Redondeo hacia Arriba
CEILING([Sales])

// Absoluto
ABS([Sales])

// Parte Entera de un Número
INT([Sales])

// Parte Decimal de un Número
[Sales] - INT([Sales])

// Exponencial
EXP([Sales])

// Logaritmo Natural
LN([Sales])

// Logaritmo Base 10
LOG([Sales])

// Potencia
POWER([Sales], 2)

// Raíz Cuadrada
SQRT([Sales])

// Número Pi
PI()

// Seno
SIN([Angle])

// Coseno
COS([Angle])

// Tangente
TAN([Angle])

// Arcotangente
ATAN([Ratio])

// Signo de un Número
SIGN([Sales])

// Valor Aleatorio entre 0 y 1
RANDOM()

// Formatear como Porcentaje
STR([Sales] * 100) + "%"

// Formatear con Decimales
STR(ROUND([Sales], 2))

// Años entre dos Fechas
DATEDIFF('year', [Order Date], TODAY())

// Mes de una Fecha
MONTH([Order Date])

// Día de una Fecha
DAY([Order Date])

// Semana del Año
WEEK([Order Date])

// Nombre del Mes
DATENAME('month', [Order Date])

// Nombre del Día de la Semana
DATENAME('weekday', [Order Date])

// Primera Letra en Mayúscula
UPPER(LEFT([Customer Name], 1)) + LOWER(MID([Customer Name], 2))

// Extraer Dominio de un Correo Electrónico
MID([Email], FIND([Email], "@") + 1, LEN([Email]) - FIND([Email], "@"))

// Contar Palabras en una Cadena
LEN(SPLIT([Customer Name], " "))

// Buscar y Reemplazar con una Expresión Regular
REGEXP_REPLACE([Customer Name], "Smith", "Jones")

// Verificar si Coincide una Expresión Regular
IF REGEXP_MATCH([Customer Name], "^S.*h$") THEN "Match" ELSE "No Match" END

// Extraer Subcadena con Expresión Regular
REGEXP_EXTRACT([Customer Name], "^[A-Za-z]+")

```
